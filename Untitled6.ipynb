```python
#we can made a function for upper two cells code plus some extra features

def immages_in_class(path_dir):
  immage_directory=[]#Making the directory of every folder in test set(We may or may not require in future)
  class_dict={}#Dict to save the number of immages in each class in test set
  immages_class_dict={}#dictionary of immages files in different classis
  #folder_directory=[]
  count=0
  for folder in os.listdir(path_dir):
    fold_direct=path_dir + folder
    #folder_directory.append(fold_direct)
    class_dict[folder]=len(os.listdir(fold_direct))
    immages=[]
    for immage in os.listdir(fold_direct):
      count+=1
      image_direct=fold_direct+"/"+immage
      immage_directory.append(image_direct)
      immages.append(immage)
    immages_class_dict[folder]=immages

  return class_dict,immages_class_dict,immage_directory,count
```

Loading the pre-trained weights of VGG16 CNN architecture. This architecture is the first runner-up in the 2014 ImageNet Challenge. It has total sixteen layers with thirteen convolutional layers. Visual Graphics Group designed it at Oxford University. We are going to usr feature extraction method as our ImageNet dataset having only ten classes is subset of orignal ImageNet dataset having thousand classes. We can also use same method or other methods of transfer learning for datasets that are differnt(not the subset of orignal dataset) and we can also use other methods of transfer learning for this dataset.

Two ways of transfer learning are.

i) Fine tuning: In this proceedure we fine tune the pre-trained model. We keep already learnt weights as initial parameters.

ii) Feature Extraction: We extract features from pre-trained model and then training a classifier on top of it.


```python
from tensorflow.keras.applications import VGG16
#we will mention "include_top=False" to remove the last three fully connected layers(including softmax layer). We have only ten classes in our model
model_pretrained = VGG16(include_top=False, weights='imagenet',input_shape=(224,224,3))#it retained the weights of 'imagenet' data. we put weights= None than it will forget all learnt features and weights will initialize randomly
model_pretrained.summary()
```

```python
# extracting useful features from train and validation dataset (featurse that VGG16 already knows from our datasetâ€™s images)
VGG16_features_train = model_pretrained.predict(train)
VGG16_features_test = model_pretrained.predict(test)
```

```python
train.labels
```

```python
test.labels
```

```python
#Encoding the labels
from tensorflow.keras.utils import to_categorical
targets_train = to_categorical(train.labels)
targets_test = to_categorical(test.labels)
print(targets_train[1])
print(targets_test[1])
```

Now we are using extracted features to train the model.

```python
#We are adding some layers at the end of model(we have removed flattening layer and onward from orignal model)

model = Sequential()

#Flattening Layer
model.add(Flatten(input_shape=(7,7,512)))#The output of the max-pooling layer has a volume shape of 7 x 7 x 512 in orignal VGG16 model.for more details Check the original VGG16 network architecture.

#Dense layer
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.4))
model.add(BatchNormalization())

#Softmax layer
model.add(Dense(10, activation='softmax'))#We have ten classes
```

```python
# compiling model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.summary()
```

```python
from keras.callbacks import EarlyStopping
#To avoid overfitting early stopping is used
earlystop = EarlyStopping(monitor = 'val_loss',
                          min_delta = 0,
                          patience = 3,
                          verbose = 1,
                          restore_best_weights = True)
callbacks=[earlystop]
```

```python

# fitting on data for 30 epochs
history=model.fit(VGG16_features_train, targets_train, epochs=50, batch_size=128, validation_data=(VGG16_features_test, targets_test),callbacks=callbacks)
```

```python
#saving model
from tensorflow.keras.models import model_from_json
model_json = model.to_json()
with open("model_final.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("model_final.h5")
```
